```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = np.linspace(0.125,1,100)

V_max = 50
beta = 1.2
p = 0.5
A = 0
SV_lottery = (p-beta*A/2)*(V_max**alpha)
SV_safe = SV_lottery/2
SV_p4 = SV_lottery
SV_p2 = 0.75*SV_lottery
SV_m2 = 0.25*SV_lottery
SV_m4 = np.zeros(SV_lottery.shape)

SV_delta = SV_lottery - SV_safe

plt.plot(alpha,SV_p4-SV_safe,'r')
plt.plot(alpha,SV_p2-SV_safe,'g')
plt.plot(alpha,SV_safe-SV_safe,'b')
plt.plot(alpha,SV_m2-SV_safe,'c')
plt.plot(alpha,SV_m4-SV_safe,'y')
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$\Delta$ SV')
```

```python
v_p4 = np.max(V_max,np.exp(SV_p4/(p*alpha)))
v_m4 = np.exp(SV_m4/(p*alpha))

plt.plot(alpha,v_p4,'r')
plt.plot(alpha,v_m4,'y')

```
